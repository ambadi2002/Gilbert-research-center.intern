**HYPOTHESIS TESTING**



*   **Z-TEST**





*   since, the sample size is greater than 30,we are executing one tail z-test







*  alpha = 0.05
*  z_critical = 1.645





*   Ho : There no significant relationship b/w year and Cyclonic Disturbances - TOTAL
*   H1 : There a significant relationship b/w year and Cyclonic Disturbances - TOTAL



```python
alpha = 0.05
z_critical = 1.645
```

**finding test statistic value**

```python
from statsmodels.stats.weightstats import ztest
```

```python
z_stat,p_value=ztest(data['Year'],data['Cyclonic Disturbances - TOTAL'])
print("z_stat :",z_stat)
```

```python
if z_stat>z_critical:
  print("Reject null hypothesis")
else:
  print("Accept null hypothesis")
```



*   There a significant relationship b/w year and Cyclonic Disturbances - TOTAL
